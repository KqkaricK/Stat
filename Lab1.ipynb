```python
# можете использовать следующие значения
# либо выбрать свои данные, а также функции (cos/sin и т д)
N = 100

x = np.arange(N)

y_1 = np.sin(x / 10)
y_2 = np.sin(x / 5)
```

Подсказка для поиска координаты текста над графиком:

min_ylim, max_ylim = plt.ylim()

```python
# Ваш код
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu1 = 0
sigma1 = 1
mu2 = 2
sigma2 = 1.5
mu3 = -2
sigma3 = 2

x1 = np.random.normal(mu1, sigma1, 5000)
x2 = np.random.normal(mu2, sigma2, 5000)
x3 = np.random.normal(mu3, sigma3, 5000)

plt.hist(x1, bins=50, density=True, color='blue', alpha=0.5, label='N(0, 1)')
plt.hist(x2, bins=50, density=True, color='red', alpha=0.5, label='N(2, 1.5)')
plt.hist(x3, bins=50, density=True, color='green', alpha=0.5, label='N(-2, 2)')

plt.legend()

plt.axvline(mu1, color='blue', linestyle='--')
plt.axvline(mu2, color='red', linestyle='--')
plt.axvline(mu3, color='green', linestyle='--')

min_ylim, max_ylim = plt.ylim()
max_ylim = max_ylim + 0.01

plt.text(mu1, max_ylim, f'μ = {mu1}', color='blue', horizontalalignment='center')
plt.text(mu2, max_ylim, f'μ = {mu2}', color='red', horizontalalignment='center')
plt.text(mu3, max_ylim, f'μ = {mu3}', color='green', horizontalalignment='center')

plt.show()
```
